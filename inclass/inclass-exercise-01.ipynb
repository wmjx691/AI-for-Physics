## Separating 0's and 1's

Task: given a handwritten image, identify it as either 0 or 1 (**binary classification**)

First we need to prepare the input data. We only need 0's and 1's from the original MNIST dataset.

```python
##############################################################################
# Selecting 0's and 1's from x_train
# How many 0's and how many 1's do we have?
##############################################################################
# Please code in this cell

zero = x_train[y_train==0]
one  = x_train[y_train==1]
print("number of zero: ", zero.shape)
print("number of one : ", one.shape)

##############################################################################
```

We have 5923 samples for zero and 6742 samples for one. We want to find some **feature** that may help separating them. One example is the average pixel density (averaging the sum of grey scales numbers over the 28x28 image).

Instead of using the original range from 0 to 255, we'll first normalize it to \[0,1] (i.e. a value between 0 and 1.). Then we will calculate the mean value for each image.



```python
##############################################################################
# Normalize the grey scale to [0,1]
sample0 = zero/255.
sample1 = one/255.

print(sample0.shape, sample1.shape)
# Calculating the mean value of pixel grey scale of each image in sample0 and sample 1
# Hint: take the mean over axis-1 and 2
# mean0 = 
##############################################################################
# Please code in this cell

mean0 = np.mean(sample0,axis=(1,2))
mean1 = np.mean(sample1,axis=(1,2))

print(mean0,mean1)

##############################################################################
```

Now we can plot the distributions of the average pixel density for 0 and 1 (mean0 and mean1), and see if it is useful for separating 0 and 1.

```python
fig = plt.figure(figsize=(6,6), dpi=80)
plt.hist(mean0, bins=50, color='y', label='True 0')
plt.hist(mean1, bins=50, color='g', alpha=0.8, label='True 1')
plt.legend()
plt.show()
```

So indeed it is a good chance that we can tell if an image is 0 or 1 by calculating its average pixel density! This tells us the importance of **feature extraction**!
